```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data_to_load_df = pd.read_csv(city_data_to_load)
ride_data_to_load_df = pd.read_csv(ride_data_to_load)
ride_data_to_load_df.head()
# Combine the data into a single dataset
merge_table = pd.merge( ride_data_to_load_df, city_data_to_load_df, on="city")


# Display the data table for preview
merge_table.head()
#merge_table.columns
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
group = merge_table.groupby(['type','city']).agg({'fare': 'mean','ride_id': 'count','driver_count':sum})

group.head()
```

```python
city_type = 'Rural'
plt.scatter(group.loc[city_type,'ride_id'], group.loc[city_type,'fare'], marker="o", facecolors="gold", alpha=0.7, edgecolors="black", s= group.loc[city_type,'driver_count'])
city_type = 'Suburban'
plt.scatter(group.loc[city_type,'ride_id'], group.loc[city_type,'fare'], marker="o", facecolors="lightskyblue", alpha=0.6, edgecolors="black", s= group.loc[city_type,'driver_count'])
city_type = 'Urban'
plt.scatter(group.loc[city_type,'ride_id'], group.loc[city_type,'fare'], marker="o",  facecolors="lightcoral", alpha=0.3, edgecolors="black", s= group.loc[city_type,'driver_count'])

plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")


plt.legend(["Rural","Suburban","Urban"], loc="best", title='City Types', scatterpoints=1)
plt.show()
```

## Total Fares by City Type

```python
#Calculate Type Percents
piechart_group = merge_table.groupby(['type','city']).agg({'fare': 'sum','ride_id': 'count','driver_count': 'sum'})
piechart_group

#Build Pie Chart
plt.title("% of Total Rides by City Type")
sizes = [piechart_group.loc['Rural','fare'].sum(),piechart_group.loc['Suburban','fare'].sum(),piechart_group.loc['Urban','fare'].sum()]
labels = ["Rural", "Suburban", "Urban"]
colors = ["gold", "lightskyblue","lightcoral"]
explode = (0, 0, 0.1)
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=170)

#Save Figure
plt.savefig("../Images/TotalFarebycity.png")

#Show Figure
plt.show()
```
